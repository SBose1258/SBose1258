LOAD DATA   

```python
!pip install pandas
import pandas as pd 
df=pd.read_csv(r'C:\Users\Subha Bose\.vscode\vs code files\Projects\ML Program 1\delaney_solubility_with_descriptors.csv')
df
```

Data Separation into x and y  

```python
y=df['logS']
y
```

```python
x= df.drop('logS', axis=1) #axis=1 is x is treated as Column axis=0 x is treated as row
x
```

Splitting the Data into Training and Testing

```python
!pip install scikit-learn

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
#Training Set: Trains the model on thsi data
x_train
```

```python
#Testing Set: tests the model that we trained on the unknown data
x_test
```

```python
y_train
```

```python
y_test
```

MODEL BUILDING

LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression

lr= LinearRegression()
lr.fit(x_train,y_train)
```

Applying the model to make a prediction

```python
y_lr_train_pred=lr.predict(x_train)
y_lr_test_pred=lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse=mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2= r2_score(y_train,y_lr_train_pred)


lr_test_mse=mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)

print('LR MSE (Train)',lr_train_mse)
print('LR R2 (Train)',lr_train_r2)
print('LR MSE (Test)',lr_test_mse)
print('LR R2 (Test)',lr_test_r2)

```

```python
lr_results=pd.DataFrame(['LinearRegression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_results.columns=['Method','Training MSE','Training R2','Test MSE','Test R2']
lr_results
```

```python
RANDOM FOREST
```

```python
from sklearn.ensemble import RandomForestRegressor
#Here we are using RandomForest Regressor Model as the y column which we are using is quantitative and we are building regression model
#if the y column was categorical then we would have used classification model
#y=quantitative=regression model=RandomForestRegressor 
#y=categorical=classification=RandomForestClassifier

rf=RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train,y_train)
```

Applying the model to make a prediction

```python
y_rf_train_pred=rf.predict(x_train)
y_rf_test_pred=rf.predict(x_test)
```

Evaluate Model Performance

```python
rf_train_mse=mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2= r2_score(y_train,y_rf_train_pred)


rf_test_mse=mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2=r2_score(y_test,y_rf_test_pred)
```

```python
rf_results=pd.DataFrame(['RandomForest',rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()
rf_results.columns=['Method','Training MSE','Training R2','Test MSE','Test R2']
rf_results
```
